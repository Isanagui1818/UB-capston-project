```python
# Creación de sample 10.000 casos solo 2024
import os
import pandas as pd

# Definir la carpeta origen donde están los archivos CSV
carpeta_origen = "C:/Users/TITAN OSCURO/Desktop/Datos bicing"  # Cambia esto por la ruta correcta

# Crear una lista para almacenar los dataframes del año 2024
dataframes_2024 = []

# Iterar sobre los archivos en la carpeta origen
for archivo in os.listdir(carpeta_origen):
    # Verificar si el archivo corresponde al año 2024 y es un CSV
    if archivo.startswith("2024") and archivo.endswith(".csv"):
        # Leer el archivo CSV y añadirlo a la lista de dataframes
        ruta_archivo = os.path.join(carpeta_origen, archivo)
        df = pd.read_csv(ruta_archivo)
        dataframes_2024.append(df)

# Combinar todos los dataframes del año 2024 en uno solo
if dataframes_2024:  # Verificar que la lista no esté vacía
    dataframe_2024 = pd.concat(dataframes_2024, ignore_index=True)

    # Crear un sample de hasta 10,000 filas
    dataframe_sample_2024 = dataframe_2024.sample(n=min(10000, len(dataframe_2024)), random_state=42)

    print("Se ha creado un sample de hasta 10,000 filas del año 2024, llamado 'dataframe_sample_2024'.")
else:
    print("No se encontraron archivos para el año 2024.")
```

```python
# eliminacoión de columnas sin información
dataframe_sample_2024.drop(columns=['traffic','V1'], inplace=True)
```

```python
# eliminación de NA por si se han colado con la selección de datos random:
dataframe_sample_2024.dropna(inplace=True)

# transformación de la variable tiempo:
dataframe_sample_2024['last_reported']=pd.to_datetime(dataframe_sample_2024['last_reported'], unit='s')
dataframe_sample_2024['last_updated']=pd.to_datetime(dataframe_sample_2024['last_updated'], unit='s')
```
